### 6 task

```python
M = np.floor(10 * np.random.random((4,5)))
print("M\n", M)

for i in range(M.shape[1]):
    M[:, i] -= np.mean(M[:, i])

print("After substraction\n", M)

for i in range(M.shape[1]):
    M[:, i] /= np.sqrt(np.sum(np.square(M[:, i]), axis=0))

print("After division\n", M)
```

```python
import numpy as np
n = 6
m = 5
a = np.random.random((n,m))
b = np.random.random((n,m))
c = np.zeros((m, m))
print('1я матрица n*m\n', a)
print()
print('2я матрица n*m\n', b)
print()
print('Матрица m*m всевозможных скалярных произведений столбцов a и b')
for i in range(m):
    for j in range(m):
        c[i, j] = np.dot(a.transpose()[i][:],  b.transpose()[j][:])  
print(c)
```

### 3 task

```python
def diff1(f, x0, h):
    return (f(x0 + h) - f(x0))/h
def diff2(f, x0, h):
    return (f(x0 + h) - f(x0 - h))/(2*h)

hs = np.logspace(-16,0, num = 50)
err1 = []
err2 = []
for h in hs:
    err1.append(abs(diff1(np.sin, 1, h) - np.cos(1)))
    err2.append(abs(diff2(np.sin, 1, h) - np.cos(1)))
    
M2 = M3 = 1

plt.figure(figsize = (18, 10))

plt.loglog(hs, err1, 'r-', ms = 16)
plt.loglog(hs, err2, 'k-', ms = 16)
plt.loglog(hs, hs*M2/2, 'r-', label = "first order", lw = 3)
plt.loglog(hs, hs**2*M3/6, 'k-', label = "first order", lw = 3)
plt.ylim(1e-14, 1)
plt.ylabel('$\Delta$')
plt.xlabel('Step')
plt.grid()
plt.show()
```
