# Задача 2. Интерполяция полиномом Лагранжа.



Допишите класс, который конструирует интерполяционный полином Лагранжа, проходящий через точки, заданные как `xk` и `yk`.

```python
import numpy as np

class LagrangeInterpolator:
    """Lagrange interpolating polynomial.
    
    Given a set of pairs ``(x_k, y_k)``, construct 
    a Lagrange polynomial ``f(x)``, such that
    
    .. math::

        f(x_k) = y_k   for k =0, ..., n-1
    
    Parameters
    ----------
    xk : array_like, shape(n,)
        Abscissas
    yk : array_like, shape(n,)
        Ordinates
    
    Attributes
    ----------
    __call__
    
    """
    def __init__(self, xk, yk):
        self.xk = np.asarray(xk, dtype=float)
        self.yk = np.asarray(yk, dtype=float)
        
    def __call__(self, *x):
        """Evaluate the interpolator at a given point.
        
        Parameters
        ----------
        x : float
        
        Returns
        -------
        the value of the interpolator at ``x``.
        """
        # YOUR CODE HERE
#         for i in range(len(self.xk)):
#             if (x == self.xk[i]):
#                 return self.yk[i]
        L = 0
        for i in range(len(self.yk)):
            l = 1
            for j in range(len(self.xk)):
                if (i != j):
                    l *= (x - self.xk[j])/(self.xk[i] - self.xk[j])
            L += l*self.yk[i]
        return L
        raise NotImplementedError()
```

```python
def runge_func(x, a=25):
    return 1.0 / (1.0 + a*x**2)

xx = np.linspace(-2, 2, 21)
yy = runge_func(xx)

lagr = LagrangeInterpolator(xx, yy)

from numpy.testing import assert_allclose

assert_allclose(yy,
                [lagr(xval) for xval in xx],
                atol=1e-14)
```

```python
# YOUR CODE AND COMMENTS HERE
import numpy as np
import matplotlib.pyplot as plt

list_m = [3,5,7,11]

def f(x):
    return 1/(1 + 25*x**2)

def printFunction(m = 3):
    x = np.linspace(-2, 2, m)
    y = [f(i) for i in x]
    plt.scatter(x, y)
    p = LagrangeInterpolator(x, y)
    plt.plot(np.linspace(-2, 2, 1001), [p(i) for i in np.linspace(-2, 2, 1001)], label = f'полином Лагранжа {m}го порядка')

plt.figure(figsize = (12, 9))
for i in list_m:
    printFunction(i)
plt.plot(np.linspace(-2, 2, 1001), [f(i) for i in np.linspace(-2, 2, 1001)], 'k-', label = 'y=1/(1+25x^2)')
plt.legend()
plt.title('Интерполяция функции Рунге полиномами Лагранжа степенями m')
plt.xlabel('x'); plt.ylabel('y')
plt.grid()
plt.show()
```

Допишите функцию, которая возвращает узлы Чебышева. 

```python
def cheb_nodes(n, a=-1, b=1):
    r"""Chebyshev nodes of degree $n$ on $[a, b]$
    """
    # YOUR CODE HERE
    y = [(a+b)/2 + ((b-a)/2) * np.cos(np.pi*(2*k-1)/(2*n)) for k in range(1, n+1)]
    return y[::-1]
    raise NotImplementedError()
```

```python
nodes_11 = cheb_nodes(11)
nodes_11 = np.asarray(nodes_11)
assert (nodes_11[1:] > nodes_11[:-1]).all()

from scipy.special import roots_chebyt
nodes, weights = roots_chebyt(5) 

assert_allclose(cheb_nodes(5),
                nodes, atol=1e-14)

assert_allclose(cheb_nodes(5, a=-1, b=3),
                nodes*2 + 1, atol=1e-14)
```

# Задача 4. Полиномы Чебышева.

Постройте интерполяцию функций

$$
y_{1}(x)=\sin (6 x)+\sin \left(60 e^{x}\right), y_{2}(x)=\frac{1}{1+1000(x+0.5)^{2}}+\frac{1}{\sqrt{1+1000(x-0.5)^{2}}}
$$

на отрезке $[-1,1]$ полиномами Чебышева. Постройте зависимость ошибки приближения от количества узлов. Сколько узлов нужно удержать в каждом из этих случаев для получения достаточно точного приближения?

**Ответ**: Для 1й функции достаточно значений в диапазоне [80, 100], тогда как для 2й функции необходимо число узлов около 100. Больше не рекомендуется, т.к. возникают бурные релаксации вблизи локальных экстремумов.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import chebyshev as M

def f1(x):
    return np.sin(6*x) + np.sin(60*np.e**x) 

def f2(x):
    return 1/(1 + 1000*(x + 0.5)**2) + 1/np.sqrt(1 + 1000*(x - 0.5)**2)

def printInt(nodes, k = 7, f = f2):
    x = cheb_nodes(nodes, a=-1, b=1)
    y = np.array([f(i) for i in x])    
    cheb = M.Chebyshev.fit(x, y, deg = k)
    plt.plot(x, y, '.')
    plt.plot(np.linspace(-1,1,1001), cheb(np.linspace(-1,1,1001)), '-', label = f'{nodes =}')
    return sum([abs(cheb(i) - f(i)) for i in np.linspace(-1,1,1001)])
```

```python
plt.subplot(1,2,2)
plt.figure(figsize = (16, 9))
list_nodes = list(range(3, 102, 6))
errors = list()
for nodes in list_nodes:
    errors.append(printInt(nodes, k = nodes, f = f2))

plt.plot(np.linspace(-1,1,1001), [f2(i) for i in np.linspace(-1,1,1001)], 'k')
plt.xlabel('x'); plt.ylabel('y')
plt.title('Интерполяция y2(x) полиномами Чебышева')
plt.legend()
plt.grid()
plt.show()

plt.subplot(2,2,2)
plt.figure(figsize = (12, 8))
plt.plot(list_nodes, errors, 'k')
plt.xlabel('nodes'); plt.ylabel('error')
# plt.legend()
plt.title('График ошибки приближения error от количества узлов nodes')
plt.grid()
plt.show()
```

```python
plt.subplot(1,2,2)
plt.figure(figsize = (16, 9))
list_nodes = list(range(5, 36, 5))
errors = list()
for nodes in list_nodes:
    errors.append(printInt(nodes, k = nodes, f = f1))

plt.plot(np.linspace(-1,1,1001), [f1(i) for i in np.linspace(-1,1,1001)], 'k')
plt.xlabel('x'); plt.ylabel('y')
plt.title('Интерполяция y1(x) полиномами Чебышева')
plt.legend()
plt.grid()
plt.show()

plt.subplot(2,2,2)
plt.figure(figsize = (12, 8))
plt.plot(list_nodes, errors, 'k')
plt.xlabel('nodes'); plt.ylabel('error')
# plt.legend()
plt.title('График ошибки приближения error от количества узлов nodes')
plt.grid()
plt.show()
```

# Задача 5. Приближение разными функциями.

Векторы х4 и у4 из архива дают значения некоторой функции на равномерной сетке на отрезке [-0.75, 0.75]. Используя scipy.optimize.curve_fit, постройте интерполяцию функции $y(x)$ функциями вида

$$
y_{1}(x)=p_{0}+p_{1} x+p_{2} x^{2}+p_{3} x^{3}+p_{4} x^{4}+p_{5} x^{5}, y_{2}(x)=\frac{p_{0}+p_{1} x+p_{2} x^{2}+p_{3} x^{3}}{1+q_{1} x+q_{2} x^{2}}
$$


Какая из этих формул лучше описывает $y(x)$ на заданном интервале? Постройте график экстраполяции $y(x)$ за пределы исходных данных. Можете ли вы угадать аналитическое выражение для $y(x) ?$

**Ответ**: Скорее всего, это функция тангенса или какая-либо совокупность полиномов нечетной степени.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

with np.load('data_interp.npz') as data:
    x, y = data['x4'], data['y4']

def f1(x, p0, p1, p2, p3, p4, p5):
    return p0 + p1*x + p2*x**2 + p3*x**3 + p4*x**4 + p5*x**5

def f2(x, p0, p1, p2, p3, q1, q2):
    return (p0 + p1*x + p2*x**2 + p3*x**3)/(1 + q1*x + q2*x**2)

plt.figure(figsize = (12, 9))
plt.plot(x, y, 'b.', label = 'data')

param, param_cov = curve_fit(f1, x, y)
plt.plot(x, [f1(i, *param) for i in x], 'r-', label = 'f1')
plt.plot(np.linspace(-2, 2, 1001), [f1(i, *param) for i in np.linspace(-2, 2, 1001)], 'r-', label = 'f1 продолжение')

param, param_cov = curve_fit(f2, x, y)
plt.plot(x, [f2(i, *param) for i in x], 'g-', label = 'f2')
plt.plot(np.linspace(-2, 2, 1001), [f2(i, *param) for i in np.linspace(-2, 2, 1001)], 'g-', label = 'f2 продолжение')

plt.title('Data')
plt.xlabel('x'); plt.ylabel('y') 
plt.legend()
plt.grid()
plt.show()
```
