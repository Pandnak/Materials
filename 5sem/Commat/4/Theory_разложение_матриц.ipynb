## LU-разложение

Теорема: 

Пусть все угловые миноры матрицы отличны от нуля, тогда матрицу $A$ можно единственным способом представить в виде произведения $A = LU$, где $L$ - невырожденная нижнетреугольная матрица, а $U$ - невырожденная верхнетреугольная матрица. 


Докажем это.


1. Для единичной матрицы очевидно
2. Далее индукцией


Любая невырожденная матрица имеет матрицы перестановок, следовательно существует перестановка $P$ такая, что $PA$ имеет ненулевые миноры и $PA = LU$.


Тогда к системе $PA = Pf$ можно применить метод Гаусса. 

##Разложение Холецкого



Для любой симметричной положительно-определенной матрицы $A$ существует единственное представление в виде $A = LL^\top$, где $L$ - нижнетреугольная матрица со строго положительными элементами на диагонали. (Для комплекснозначных матриц $A = LL^*$)


Матрицы такого вида возникают при использовании метода наименьших квадратов и численном решении дифференциальных уравнений.

Данный метод более устойчив по сравнению с методом Гаусса или $LU$-разложением и более быстрый. 

## Сингулярное разложение.

Теорема 

Любая матрица $A \in \mathbb{C}^{n \times m}$ может быть представлена в виде произведения трёх матриц:

$$
A=U \Sigma V^*
$$

где $U$ - унитарная матрица $n \times K, V-$ унитарная матрица $m \times K, K=\min (m, n)$,
$\Sigma$ - диагональная матрица с неотрицательными элементами на диагонали $\sigma_1 \geq \ldots, \geq \sigma_K$.
Более того, если $\operatorname{rank}(A)=r$, тогда $\sigma_{r+1}=\cdots=\sigma_K=0$.

Доказательство

- Матрица $A^* A$ эрмитова, поэтому диагонализуема в унитарном базисе (подробнее будет рассказано на следующих лекциях).
- $A^* A \succeq 0$ (неотрицательно определена), поэтому собственные значения неотрицательны. Поэтому существует унитарная матрица $V=\left[v_1, \ldots, v_n\right]$ такая чाо
$$
V^* A^* A V=\operatorname{diag}\left(\sigma_1^2, \ldots, \sigma_n^2\right), \quad \sigma_1 \geq \sigma_2 \geq \cdots \geq \sigma_n \geq 0
$$
Пусть $\sigma_i=0$ для $i>r$, где $r$ некоторое целое число.
Пусть $V_r=\left[v_1, \ldots, v_r\right], \Sigma_r=\operatorname{diag}\left(\sigma_1, \ldots, \sigma_r\right)$. Таким образом,
$$
V_r^* A^* A V_r=\Sigma_r^2 \quad \Longrightarrow \quad\left(\Sigma_r^{-1} V_r^* A^*\right)\left(A V_r \Sigma_r^{-1}\right)=I .
$$
В результате для матрицы $U_r=A V_r \Sigma_r^{-1}$ выполнено $U_r^* U_r=I$ и следовательно её столбцы ортогональны.
Присоединим к матрице $U_r$ произвольные ортогоналоьные столбцы, которые также ортогональны столбцам $U_r$ и обозначим такую матрицу $U$. Тогда
$$
A V=U\left[\begin{array}{cc}
\Sigma_r & 0 \\
0 & 0
\end{array}\right] \quad \Longrightarrow \quad U^* A V=\left[\begin{array}{cc}
\Sigma_r & 0 \\
0 & 0
\end{array}\right]
$$
Поскольку умножение на невырожденную матрицу не меняет ранга $A$, получим $r=\operatorname{rank}(A)$.

$\bf{Теорема:}$ 

Пусть $r<\operatorname{rank}(A), A_r=U_r \Sigma_r V_r^*$. 
Тогда

$$
\min _{\operatorname{rank}(B)=r}\|A-B\|_2=\left\|A-A_r\right\|_2=\sigma_{r+1} .
$$

Похожий результат справедлив и для $\|\cdot\|_{F,}$ но $\left\|A-A_r\right\|_F=\sqrt{\sigma_{r+1}^2+\cdots+\sigma_{\min (n, m)}^2}$.

Доказательство

- Поскольку $\operatorname{rank}(B)=r$, то справедливо равенство $\operatorname{dim} \operatorname{ker} B=n-r$.
- Тогда существует $z \neq 0$ такой что $z \in \operatorname{ker}(B) \cap \mathcal{L}\left(v_1, \ldots, v_{r+1}\right)$ (так как $\operatorname{dim}\left\{v_1, \ldots, v_{r+1}\right\}=r+1$.
- Зафиксируем $\|z\|=1$. Тогда,
$$
\|A-B\|_2^2 \geq\|(A-B) z\|_2^2=\|A z\|_2^2=\left\|U \Sigma V^* z\right\|_2=\left\|\Sigma V^* z\right\|_2=\sum_{i=1}^n \sigma_i^2\left(v_i^* z\right)^2=\sum_{i=1}^{r+1} \sigma_i^2\left(v_i^* z\right)^2 \geq \sigma_{r+1}^2 \sum_{i=1}^{r+1}\left(v_i^* z\right)^2=\sigma_{r+1}^2
$$
так как $\sigma_1 \geq \cdots \geq \sigma_{r+1} и$
$$
\sum_{i=1}^{r+1}\left(v_i^* z\right)^2=\|V z\|_2^2=\|z\|_2^2=1 .
$$


Основной результат о малоранговой аппроксимации
$$
\min _{A_r}\left\|A-A_r\right\|_2=\sigma_{r+1}, \quad \min _{A_r}\left\|A-A_r\right\|_F=\sqrt{\sigma_{r+1}^2+\cdots+\sigma_K^2}
$$
поэтому так важно смотреть на степень убывания сингулярных чисел матрицы.

```python
#Computing matrix rank
import numpy as np
n = 50 
a = np.ones((n, n))
print('Rank of the matrix:', np.linalg.matrix_rank(a))
b = a + 1e-5 * np.random.randn(n, n)
print('Rank of the matrix:', np.linalg.matrix_rank(b))
```

```python
u, s, v = np.linalg.svd(b)
print(s[1]/s[0])
r = 1
u1 = u[:, :r]
s1 = s[:r]
v1 = v[:r, :]
a1 = u1.dot(np.diag(s1).dot(v1))
print(np.linalg.norm(b - a1, 2)/s[0])
```

Мы можем использовать SVD для вычисления аппроксимации матриц, полученных из вычисления некоторой функции в точках сетки
$$
a_{i j}=f\left(x_i, y_j\right),
$$
где $f$ заданная функция, а $x_i, i=1, \ldots, n$ и $y_j, j=1, \ldots, m$ некоторые одномерные сетки.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.rc("text", usetex=False)
import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("talk")
plt.xkcd()
n = 1000
a = [[1.0/(i+j+1) for i in range(n)] for j in range(n)] #Hilbert matrix 
a = np.array(a)
u, s, v = np.linalg.svd(a)
plt.semilogy(s[:30]/s[0])
plt.ylabel(r"$\sigma_i / \sigma_0$", fontsize=24)
plt.xlabel(r"Singular value index, $i$", fontsize=24)
plt.grid()
plt.show()
```

Теперь попробуем аппроксимировать заданную функцию



```python
import numpy as np
plt.rc("text", usetex=False)
n = 128
t = np.linspace(0, 5, n)
x, y = np.meshgrid(t, t)
f = 1.0 / (x + y + 0.5) # test your own function. Check 1.0 / (x - y + 0.5)
u, s, v = np.linalg.svd(f, full_matrices=False)
r = 10
u = u[:, :r]
s = s[:r]
v = v[:r, :] # Mind the transpose here!
fappr = u.dot(np.diag(s).dot(v))
er = np.linalg.norm(fappr - f, 'fro') / np.linalg.norm(f, 'fro')
print("Approximation error =", er)
plt.semilogy(s/s[0])
plt.ylabel(r"$\sigma_i / \sigma_0$", fontsize=24)
plt.xlabel(r"Singular value index, $i$", fontsize=24)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#plt.xkcd()
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(x, y, f)
ax.set_title('Original function')
ax = fig.add_subplot(122, projection='3d')
ax.plot_surface(x, y, fappr - f)
ax.set_title('Approximation error with rank=%d, err=%3.1e' % (r, er))
fig.subplots_adjust()
fig.tight_layout()
```
