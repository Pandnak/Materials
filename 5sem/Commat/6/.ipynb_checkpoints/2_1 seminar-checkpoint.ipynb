#  Прямая с шумом

Сгенерируем точки $(x_i, y_i)$ по закону
$$
    y = w_1 x + w_0 + \epsilon,
$$
где $\epsilon$ - случайная величина из нормального распределения, моделирующая разброс точек. 

```python
# Возьмем $\epsilon \sim \mathcal{N}(0, 0.3)$. 
```

```python
true_w1 = 1.7
true_w0 = 2.3
# print('True w:', true_w1, true_w0)

xs = np.arange(1, 5, 0.2)
noise = np.random.normal(0, 0.3, size=xs.shape)
ys = xs * true_w1 + true_w0 + noise
# [x * true_w1 + true_w0 for x in xs] -- BAD!
```

```python
plt.scatter(xs, ys)
plt.plot(xs, ys - noise, c='r')
plt.show()
```

Применим формулы метода наименьших квадратов, выведенные на лекции:

$$
    \hat{w_1} = \sum_{i=1}^{l} \frac{(x_i - \bar{x})(y_i - \bar{y})} {(x_i - \bar{x})^2} = \frac{\hat{\text{cov}}(x, y)}{\hat{\text{var}} \, x}, \qquad \hat{w_0} = \bar{y} - \hat{w_1} \bar{x},
$$
где $\bar{x} = \frac{1}{l} \sum_{i=1}^{l} x_i$ и $\bar{y} = \frac{1}{l} \sum_{i=1}^{l} y_i$ - это средние по выборке.

Выборочные дисперсии и ковариация:
$$
    \hat{\text{var}} \, x = \frac{1}{l} \sum_{i=1}^{l} (x_i - \bar{x})^2 \qquad
    \hat{\text{var}} \, y = \frac{1}{l} \sum_{i=1}^{l} (y_i - \bar{y})^2 \qquad
    \hat{\text{cov}}(x, y) = \frac{1}{l} \sum_{i=1}^{l} (x_i - \bar{x})(y_i - \bar{y})
$$

image.png

```python
var_x = np.mean((xs - xs.mean()) ** 2)
var_y = np.mean((ys - ys.mean()) ** 2)
cov_xy = np.mean((xs - xs.mean()) * (ys - ys.mean()))

(var_x, var_y, cov_xy)
```

```python
(np.var(xs), np.var(ys))
```

```python
np.cov([xs, ys], bias=True)
```

```python
(np.var(xs), np.var(ys), np.cov([xs, ys], bias=True)[0, 1])
```

```python
w1 = cov_xy / var_x
w0 = ys.mean() - w1 * xs.mean()

print('True w:', true_w1, true_w0)
print('Restored w:', w1, w0)
```

```python
plt.scatter(xs, ys)
plt.plot(xs, ys - noise, c='r')
plt.plot(xs, xs * w1 + w0, c='b')
plt.show()
```

Посчитаем коэффицент линейной корреляции Пирсона:

$$
    r = \frac{\hat{\text{cov}}(x, y)}{\sqrt{ \hat{\text{var}} \, x \, \hat{\text{var}} \, y}}
$$

```python
r = cov_xy / np.sqrt(var_x * var_y)
r
```

# Псевдообратные матрицы

Псевдообратные матрицы — обобощение обратных матриц в линейной алгебре. Псевдообратная матрица к матрице A обозначается A+. Псевдообращение можно понимать как наилучшую апроксимацию (по методу наименьших квадратов) решения соответствующей системы линейных уравнений. Псевдообращение определено для любых матриц над действительными и комплексными числами 

$A^{+}$называется псевдообратной матрицей для матрицы $A$, если она удовлетворяет следующим критериям:
1. $A A^{+} A=A$;
2. $A^{+} A A^{+}=A^{+} \quad$ ( $A^{+}$является слабым обращением в мультипликативной полугруппе);
3. $\left(A A^{+}\right)^*=A A^{+} \quad$ (это означает, что $A A^{+}-$эрмитова матрица);
4. $\left(A^{+} A\right)^*=A^{+} A \quad\left(A^{+} A\right.$ - тоже эрмитова матрица).
Здесь $M^*$ - эрмитова сопряжённая матрица $M$. Для матриц над полем действительных чисел $M ^*= M^T$. 

По методу наименьших квадратов для решения несовместной СЛАУ $A x=b$, состоящей из $\mathbb{M}$ уравнений с $N$ неизвестными, необходимо решить уравнение
$$
A^T A x=A^T b,
$$
называемое нормальным уравнением. Пусть столбцы матрицы $A$ линейно независимы, тогда она обратима и система имеет единственное решение
$$
x=\left(A^T A\right)^{-1} A^T b .
$$
Таким образом мы приходим к понятию псевдообращения действительных матриц:
$$
A^{+}=\left(A^T A\right)^{-1} A^T .
$$ 

Свойства псевдообратной матрицы.

- Псевдообращение обратимо, более того, эта операция обратна самой себе:
$$
\left(A^{+}\right)^{+}=A
$$
- Псевдообращение коммутирует с транспонированием, сопряжением и эрмитовым сопряжением:
$$
\begin{aligned}
&\left(A^T\right)^{+}=\left(A^{+}\right)^T, \\
&(\bar{A})^{+}=\bar{A}^{+}
\end{aligned}
$$
$$
\left(A^*\right)^{+}=\left(A^{+}\right)^* \text {. }
$$
- Псевдообратное произведения матрицы $A$ на скаляр $\alpha$ равно соответствующему произведению матрицы $A^{+}$на обратное число $\alpha-1:$ $(\alpha A)^{+}=\alpha^{-1} A^{+}$, для $\alpha \neq 0$
- Если псевдообратная матрица для $A^* A$ уже известна, она может быть использовано для вычисления $A^{+}$:
$$
A^{+}=\left(A^* A\right)^{+} A^* \text {. }
$$
- Аналогично, если матрица $\left(A A^*\right)^{+}$уже известна:
$$
A^{+}=A^*\left(A A^*\right)^{+} .
$$

Особые случаи.

Если столбцы матрицы $A$ линейно независимы, тогда матрица $A^* A$ обратима. В таком случае псевдообратная матрица задаётся формулой
$$
A^{+}=\left(A^* A\right)^{-1} A^* .
$$
Отсюда следует что $A^{+}$- левая обратная матрица для $\mathrm{A:} A^{+} A=I$.
- Если строки матрицы $A$ линейно независимы, тогда матрица $A A^*$ обратима. В таком случае псевдообратная матрица задаётся формулой
$$
A^{+}=A^*\left(A A^*\right)^{-1} \text {. }
$$
Отсюда следует, что $A^{+}-$правая обратная матрица для $A: A A^{+}=I$.
- Если и столбцы и строки линейно независимы (что верно для квадратных невырожденных матриц), псевдообращение равно обращению:
$$
A^{+}=A^{-1} \text {. }
$$
- Если и и $B$ таковы, что произведение $A B$ определено, и
- либо $A^* A=I$,
- либо $B B^*=I$,
" либо столбцы $A$ линейно независимы и строки $B$ линейно независимы, тогда $(A B)^{+}=B^{+} A^{+}$.
- Псевдообращение можно применять и к скалярам, и к векторам. Псевдообратный к скаляру $x-$ ноль, если $x-$ ноль, и обратный к $x$ в противном случае:
$$
x^{+}=\left\{\begin{array}{cc}
0, & x=0 ; \\
x^{-1}, & x \neq 0 .
\end{array}\right.
$$
своей длины:
$$
x^{+}= \begin{cases}0^T, & x=0 ; \\ \frac{x^*}{x^* x}, & x \neq 0 .\end{cases}
$$
Для доказательства достаточно проверить, что эти величины удовлетворяют определению псевдообратных. 

Воспользуемся формализмом матрицы объект-признак и решением МНК через псевдообратную матрицу.

# Геометрическая интерпретация МНК.

Чтобы перейти к геометрической интерпретации МНК, сначала надо понять интуитивное содержание метода наименьших квадратов. Цель МНК такова: мы должны подобрать такие $p_o$ и $p_1$ чтобы минимизировать квадрат длины вектора $\mathbf{e} = \mathbf{Y}- \hat{\mathbf{Y}}$ (поскольку квадрат длины этого вектора как раз есть сумма квадратов остатков $)$, где $\hat{\mathbf{Y}}=p_0+p_1X$, вектор $Y$ лежит в пространстве, которое натянуто на единичный вектор 1 и вектор X. Это соответствует ортогональности векторов е и Y, т.е. Y является ортогональной проекцией вектора Y на плоскость, образованную векторами 1 и X . 

image.png

Вектор $\overline{\mathbf{Y}}$ является ортогональной проекцией вектора $\mathbf{Y}$ на единичный вектор 1. Тогда по теореме о трех перпендикулярах получим, что проекция вектора $\hat{\mathbf{Y}}$ на вектор 1 совпадает с $\overline{\mathbf{Y}}$.

Рассмотрим прямоугольный треугольник со сторонами $\hat{\mathbf{Y}}-\overline{\mathbf{Y}}, \mathbf{Y}-\overline{\mathbf{Y}}$ и $\mathbf{e}$. По теореме Пифагора получим равенство $\|\mathbf{Y}-\overline{\mathbf{Y}}\|^2=\|\hat{\mathbf{Y}}-\overline{\mathbf{Y}}\|^2+\|\mathbf{e}\|^2$, что совпадает с равенством TSS $=E S S+$ RSS . Также $R^2=\frac{E S S}{T S S}$, причем известно, что косинус угла в прямоугольном треугольнике равен отношению длины прилежащего катета к длине гипотенузы. Соответственно, $R^2=\cos ^2 \varphi$, где $\varphi-$ угол между сторонами $\hat{\mathbf{Y}}-\overline{\mathbf{Y}}$ и $\mathbf{Y}-\overline{\mathbf{Y}}$.

# Коэффициент детерминации.


Коэффициент детерминации (R^2 - R-квадрат) — это доля дисперсии зависимой переменной, объясняемая рассматриваемой моделью. 

Истинный коэффициент детерминации модели зависимости случайной величины $y$ от признаков $x$ определяется следующим образом:

$$
R^2=1-\frac{V(y \mid x)}{V(y)}=1-\frac{\sigma^2}{\sigma_y^2},
$$

где $V(y \mid x)=\sigma^2-$ условная (по признакам $x$ ) дисперсия зависимой переменной (дисперсия случайной ошибки модели).

В данном определении используются истинные параметры, характеризующие распределение случайных величин. Если использовать выборочную оценку значений соответствующих дисперсий, то получим формулу для выборочного коэффициента детерминации (который обычно и подразумевается под коэффициентом детерминации):

$$
R^2=1-\frac{\widehat{\sigma}^2}{\widehat{\sigma}_y^2}=1-\frac{R S S / n}{T S S / n}=1-\frac{R S S}{T S S},
$$

где

RSS $=\sum_{t=1}^n e_t^2=\sum_{t=1}^n\left(y_t-\widehat{y}_t\right)^2-$ сумма квадратов

 $T S S=\sum_{t=1}^n\left(y_t-\bar{y}\right)^2=n \widehat{\sigma}_y^2-$ общая дисперсия,


$y_t, \widehat{y}_t-$ соответственно, фактические и расчетные значения объясняемой переменной,


$\bar{y}=\frac{1}{n} \sum_{i=1}^n y_i-$ выборочное среднее.


В случае линейной регрессии с константой $T S S=R S S+E S S$, где $E S S=\sum_{t=1}^n\left(\widehat{y}_t-\bar{y}\right)^2-$ объяснённая сумма квадратов, поэтому получаем более простое определение в этом случае. Коэффициент детерминации - это доля объяснённой дисперсии в общей:

$$
R^2=\frac{E S S}{T S S} .
$$
